```python
def calculate_z_score(finaldata,atemp):              # detecting outlier
  atemp_mean=np.mean(finaldata["atemp"])
  atemp_std_deviation=np.std(finaldata["atemp"])
  z_score=(finaldata["atemp"]-atemp_mean)/atemp_std_deviation
  return z_score
```

```python
from scipy import stats
finaldata['z_score'] = np.abs(stats.zscore(finaldata['atemp']))

# Define a threshold for Z-score, usually 3
threshold = 3

# Find the outliers
outliers_z = finaldata[finaldata['z_score'] > threshold]
print(outliers_z)
```
